# Using Lazy Predict to quickly screen many algorithms to select candidates for hyperparameter tuning

```python
from lazypredict.Supervised import LazyClassifier
import pandas as pd
import numpy as np
```

```python
# load processed upscaled datasets. These are prepared in brotherhood_peter-12875237-week2_Experiment_2 balance the data.ipynb


X_ups_data = np.load('../data/processed/X_ups_data.npy')
X_ups_test = np.load('../data/processed/X_ups_test.npy')
y_ups_data = np.load('../data/processed/y_ups_data.npy')
y_ups_test = np.load('../data/processed/y_ups_test.npy')


print(X_ups_data.shape)
print(X_ups_test.shape)
print(y_ups_data.shape)
print(y_ups_test.shape)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_ups_data, X_ups_test, y_ups_data, y_ups_test)

#print(models)
```
